```python
from soccerplots.radar_chart import Radar
import pandas as pd
```

```python
jugadores=['Iñaki Williams','Fabian Orellana']
```

```python
df=pd.read_excel('20_03_2021_field_players_tournament_Spain.PrimeraDivision.xlsx')
df.head()
```

```python
df.dtypes
```

```python
df_limpio=df.drop(['Unnamed: 0'], axis=1)
df_limpio
```

```python
df_limpio.rename(columns={'Unnamed: 1':'Jugador'}, inplace=True)
df_limpio_index=df_limpio.set_index(['Jugador'])
df_limpio_index
```

```python
lista_columnas=list(df_limpio_index.columns)
lista_columnas
```

```python
variables = [
 '% de efectividad de pases',
 '% de efectividad de los centros',
 '% disputas ganadas',
 '% disputas por arriba ganadas',
 '% acciones efectivas',
 '% de efectividad en lanzamientos',
 '% disputas defensivas ganadas',
 '% disputas en ataque ganadas',
 'Regates con éxito, %',
 'Robos de balón con éxito, %'
]

df_limpio_variables=df_limpio_index[variables]
df_limpio_variables.head(10)
```

```python
##limpieza de porcentajes(%) y valores nulos (-)
df_clean_perc=df_limpio_variables.replace('%','',regex=True)
df_clean_score=df_clean_perc.replace('-','0',regex=True)
df_clean_score
```

```python
#convertir todo el df a numerico
df_final_num=df_clean_score.apply(pd.to_numeric)
df_final_num.dtypes
```

```python
df_final_num
```

```python
df_jugadores=df_final_num[df_final_num.index.isin(jugadores)]
df_jugadores
```

```python
valores=df_jugadores.values
valores
```

```python
ranges=df_jugadores.apply(lambda x: pd.Series([0, x.max()])).T.values.tolist()
ranges
```

```python
jugadores_radar=df_jugadores.index.values
jugadores_radar
```

```python
## parameter names
params = variables

## range values
ranges = ranges

## parameter value
values = valores

## title
title = dict(
    title_name="Visualización AFEC FA",
    subtitle_name=jugadores_radar[0],
    subtitle_color='#B6282F',
    subtitle_name_2=jugadores_radar[1],
    subtitle_color_2='#344D94',
    title_fontsize=18,
    subtitle_fontsize=15,
)

## endnote 
endnote = "RadarChart curso AFEC FA\nVisualization made by: Anmol Durgapal(@slothfulwave612)\nAll units are in per90"

## instantiate object
radar = Radar()

## plot radar -- compare
fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                           radar_color=['#B6282F', '#344D94'], 
                           title=title, endnote=endnote,
                            filename='afecfa.jpg', dpi=500, 
                           compare=True)
```
